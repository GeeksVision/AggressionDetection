```python
#algorithm 1: this calculates the immediate slope between two points and then calculates the 
#lat and long adjustment for each point

#http://www.csgnetwork.com/degreelenllavcalc.html used this site for lat long to feet conversion

def xyAmount (dataframe, i):
    y = dataframe.latitude[i+1]-dataframe.latitude[i]
    #switch for longitude because increasing longitude indicates westward
    x = dataframe.longitude[i]-dataframe.longitude[i+1]
    if x == 0:
        x += 0.0000000001
    slope = y/x
    angle = math.degrees(math.atan(slope))
    if (x < 0) & (y > 0):
        angle +=90
    elif (x < 0) & (y < 0):
        angle +=180

    x_adj = 12*math.cos(math.degrees(angle)) * 0.00000386573
    y_adj = 12*math.sin(math.degrees(angle)) * 0.00000274269

    return (angle, x_adj, y_adj)
#example
angle, long_adj, lat_adj = xyAmount(dfC, 1)
print(angle, long_adj, lat_adj)
```

```python
#adjust over 1 full lane
def adjust1Lane(dfSnap, long_adj, lat_adj):
    dfAdj = []
    for i in range(0,len(dfSnap)-1):
        dfAdj.append([dfSnap.latitude[i] + lat_adj, dfSnap.longitude[i] - long_adj])
    
    dfAdj = pd.DataFrame(dfAdj, columns = ['latitude', 'longitude'])
    return dfAdj

dfAdj = adjust1Lane(dfSnap, long_adj, lat_adj)
dfAdj
```

```python
#adjust half a lane
def adjustHalfLane(dfSnap, long_adj, lat_adj):
    dfAdj = []
    for i in range(0,len(dfSnap)-1):
        dfAdj.append([dfSnap.latitude[i] + (0.5*lat_adj), dfSnap.longitude[i] - (0.5*long_adj)])
    
    dfAdj = pd.DataFrame(dfAdj, columns = ['latitude', 'longitude'])
    return dfAdj

dfAdjHalf = adjustHalfLane(dfSnap, long_adj, lat_adj)
```

```python
#adjust 1.5 lanes
def adjustOneAndHalfLane(dfSnap, long_adj, lat_adj):
    dfAdj = []
    for i in range(0,len(dfSnap)-1):
        dfAdj.append([dfSnap.latitude[i] + (1.5*lat_adj), dfSnap.longitude[i] - (1.5*long_adj)])
    
    dfAdj = pd.DataFrame(dfAdj, columns = ['latitude', 'longitude'])
    return dfAdj

dfAdjOneAndHalf = adjustOneAndHalfLane(dfSnap, long_adj, lat_adj)
```

```python
#adjust 2.5 lanes
def adjustTwoAndHalfLane(dfSnap, long_adj, lat_adj):
    dfAdj = []
    for i in range(0,len(dfSnap)-1):
        dfAdj.append([dfSnap.latitude[i] + (2.5*lat_adj), dfSnap.longitude[i] - (2.5*long_adj)])
    
    dfAdj = pd.DataFrame(dfAdj, columns = ['latitude', 'longitude'])
    return dfAdj

dfAdjTwoAndHalf = adjustTwoAndHalfLane(dfSnap, long_adj, lat_adj)
```
